```python
# Draw a mathmetical different polynomial model.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10, 10)
```

```python
y=x**3+x**2+x+1
```

```python
plt.plot(x, y, label="y=x^3+x^2+x+1")
plt.xlabel("Independent variable")
plt.ylabel("dependent variable")
plt.legend()
plt.title("Graph of Polynomial degree 3")
plt.grid(True)
plt.show()
```

```python
y=x**2+x+1
```

```python
plt.plot(x, y, label="y=x^2+x+1")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.legend()
plt.grid()
plt.title("Graph of polynomial degree 2")
plt.show()
```

```python
y=np.exp(x)+1
plt.plot(x, y, label="y=e^x+1")
plt.legend()
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.grid()
plt.title("Graph of Exponential")
plt.show()
```

```python
x=np.linspace(1, 10) #we can't use negative number or zero in logarithm. always use +ve number
y=np.log(x)+1
plt.plot(x, y, label="y=log(x)+1")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.title("Graph of logarithic")
plt.legend()
plt.grid()
plt.show()
```

```python
x=np.linspace(-10, 10)
y=1/(1+np.exp(-x)+1)
plt.plot(x, y, label="y=1/1+e^x+1")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.title("Graph of Sigmoidal or logistic function")
plt.grid()
plt.legend()
plt.show()
```
