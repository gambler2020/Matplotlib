```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
n=np.linspace(1, 100)
```

```python
plt.plot(n, 7*n*n+6*n+5)
```

```python
#now if we add another function 20*n in the above example
```

```python
plt.plot(n, 7*n*n+6*n+5, label="7*n*n+6*n+5")
plt.plot(n, 20*n, label="20*n")
plt.legend()
```

```python
plt.plot(n, 7*n*n+6*n+5, label="7*n*n+6*n+5")
plt.plot(n, 20*n, label="20*n")
plt.plot(n, 3*n*n+7*n*n+6*n+5, label="3*n*n+7*n*n+6*n+5")
plt.legend(loc="upper left")
```

```python
# we can ommit 6*n+5 from 7*n*n+6*n+5 because thats not going to effect any on graph so lets visualise it
#(7*n*n+6*n+5) Vs (7*n*n)
```

```python
plt.plot(n, 7*n*n+6*n+5, label="7n^2+6n+5")
plt.plot(n, 7*n*n, label="7n^2")
plt.title("(7n^2+6n+5 Vs 7n^2)", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
# now plot (7n^2+6n+5)/(7n^2) and visualise it
```

```python
plt.plot(n, (7*n*n+6*n+5)/(7*n*n), label="7n^2+6n+5")
```

```python
# now above figure we can see it that as we grow from n=1 to n=100 function is going towards 1.
```

```python
# just visualise the functions y=x^3, y=x^2, y=x
```

```python
n=np.linspace(1, 10)
plt.plot(n, n*n*n, label="x^3")
plt.plot(n, n*n, label="x^2")
plt.plot(n, n, label="x")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
```

```python
#we can see that as x grows function y=x^3 grows more and more fastre than y=x^2 and y=x
```

```python
## Now as we know that Exponential function grows more faster than polynomial lets look at the graph and try to visualose it
```

```python
n=np.linspace(1, 10)
plt.plot(n, n**4, label="n^4")
plt.plot(n, 2**n, label="2^n")
plt.legend()
```

```python
## HaHaHa lol this result showing us wrong but just wait and look over n carefully we took n from 1 to 10 only
# now lets take n=1 to n=20 then we will see whats the result we are getting
```

```python
n=np.linspace(1, 20)
plt.plot(n, n**4, label="n^4")
plt.plot(n, 2**n, label="2^n")
plt.legend()
```

```python
# did you saw the magic here...lol!! so this is the correct result. polynomial always grows lesser than exponential.
```

```python
## Any poly logarithm grows slower than polynomial. lets visualise it
```

```python
n=np.linspace(1, 20)
plt.plot(n, np.log(n), label="y=log(n)")
plt.plot(n, n, label="y=n")
plt.legend()
```

```python
# we can very very easily identify which one is growing. log(n) is not even touching y=n.
```

```python
# lets make more graph and visualite it
```

```python
n = np.linspace(1, 100)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
```

```python
n = np.linspace(1, 10 ** 6)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
```

```python
n = np.linspace(1, 10 ** 8)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
```

```python
n = np.linspace(1, 10 ** 15)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
```
