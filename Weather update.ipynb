```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, linewidth=0.5, marker="*", markersize=8, color="green", label="Max_temp")
plt.plot(days, min_t, linewidth=0.5, marker="*", markersize=8, color="blue", label="Min_temp")
plt.plot(days, avg_t, linewidth=0.5, marker="*", markersize=8, color="red", label="Avg_temp")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend()
plt.title("Weather Update")
plt.grid()
```
